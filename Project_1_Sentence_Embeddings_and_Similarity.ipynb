```python
#Importing the Library
from sentence_transformers import SentenceTransformer, util
```

Loading a Pre-trained Model
To use sentence embeddings, you’ll need a pre-trained model. All Sentence Transformers models are hosted on Hugging Face.

There are over 100 models available, many of which support multi-language embeddings. These models are trained on diverse datasets and fine-tuned for specific tasks such as semantic similarity, paraphrase mining, and information retrieval.

You can load a model like this:

```python
model  = SentenceTransformer ('all-MiniLM-L6-v2')
```

This model:

Provides a good tradeoff between speed and accuracy

Has an accuracy of approximately 68% on STS benchmarks

Is much faster than heavier models

Is ideal for general-purpose usage

On first run, the model will be downloaded from Hugging Face and cached locally (e.g., in Colab). On subsequent runs, the cached version will be used.

```python
sentences = [ 'the cat sits outside','the new moview is awesome', 'the new movie is really great', 'the dog barks on strangers']
```

```python
embeddings = model.encode(sentences=sentences, convert_to_tensor=True)
```

```python
for sent,embed in zip(sentences, embeddings):
  print ("Sentence: ", sent)
  print("len(Embeddings:", len(embed))
  # print(embed)
```

###check similartiy between these sentences

```python
cosine_score = util.cos_sim(embeddings, embeddings)
```

```python
#used paraphares mining to matach the top ranked sentences from the document like 1000000

paraphrases = util.paraphrase_mining(model, sentences)
```
