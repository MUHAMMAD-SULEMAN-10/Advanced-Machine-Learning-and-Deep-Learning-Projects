```python
# Install required libraries

!pip install -q transformers
from transformers import pipeline
!pip install -q PyMuPDF
!pip install -q python-docx


# Upload your PDF
from google.colab import files
import fitz  # PyMuPDF

uploaded = files.upload()
pdf_path = list(uploaded.keys())[0]

# Extract text from PDF
def extract_text_from_pdf(file_path):
    doc = fitz.open(file_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

full_text = extract_text_from_pdf(pdf_path)
print("PDF Text Extracted.")
```

```python
from transformers import pipeline

# Initialize the summarizer
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Break text into smaller chunks (BART has a limit of ~1024 tokens)
def chunk_text(text, max_chunk_length=1000):
    sentences = text.split('. ')
    chunks = []
    chunk = ""
    for sentence in sentences:
        if len(chunk) + len(sentence) <= max_chunk_length:
            chunk += sentence + ". "
        else:
            chunks.append(chunk.strip())
            chunk = sentence + ". "
    chunks.append(chunk.strip())
    return chunks

chunks = chunk_text(full_text)
# print(f"Text split into {len(chunks)} chunks.")

# Summarize each chunk
summaries = []
for i, chunk in enumerate(chunks):
    # print(f"Summarizing chunk {i + 1} of {len(chunks)}...")
    summary = summarizer(chunk, max_length=130, min_length=3, do_sample=False)[0]['summary_text']
    summaries.append(summary)

# Combine and print the full summary
final_summary = "\n\n".join(summaries)

print("\nFinal Summary:\n")
print(final_summary)

```

```python
# ✅ Install required libraries
!pip install -q transformers PyMuPDF

# ✅ Imports
from transformers import pipeline
import fitz  # PyMuPDF
from google.colab import files

# ✅ Upload PDF
uploaded = files.upload()
pdf_path = list(uploaded.keys())[0]

# ✅ Extract text from PDF
def extract_text_from_pdf(file_path):
    doc = fitz.open(file_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

# ✅ Chunk text into smaller parts (500 words each)
def chunk_text(text, max_words=500):
    words = text.split()
    return [' '.join(words[i:i + max_words]) for i in range(0, len(words), max_words)]

# ✅ Summarize chunks using a small, fast model
def summarize_chunks(chunks):
    summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")
    summaries = []
    for chunk in chunks:
        summary = summarizer(chunk, max_length=120, min_length=30, do_sample=False)[0]['summary_text']
        summaries.append(summary)
    return summaries

# ✅ Run everything
print("Extracting text...")
full_text = extract_text_from_pdf(pdf_path)

print("Chunking text...")
chunks = chunk_text(full_text, max_words=500)

print(f"Summarizing {len(chunks)} chunks...")
summaries = summarize_chunks(chunks)

# ✅ Final summary
final_summary = "\n\n".join(summaries)
print("\n--- ✅ FINAL SUMMARY ---\n")
print(final_summary)
```

```python
# ✅ Install required libraries
!pip install -q transformers PyMuPDF tqdm

# ✅ Imports
from transformers import pipeline
import fitz  # PyMuPDF
from google.colab import files
from tqdm import tqdm
import torch

# ✅ Upload PDF
uploaded = files.upload()
pdf_path = list(uploaded.keys())[0]

# ✅ Extract text from PDF
def extract_text_from_pdf(file_path):
    doc = fitz.open(file_path)
    return " ".join([page.get_text() for page in doc])

# ✅ Chunk text into ~450-word chunks (safe for model limits)
def chunk_text(text, max_words=450):
    words = text.split()
    return [' '.join(words[i:i + max_words]) for i in range(0, len(words), max_words)]

# ✅ Summarize chunks using fast transformer model
def summarize_chunks(chunks):
    summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6", device=0 if torch.cuda.is_available() else -1)
    return [summarizer(chunk, max_length=120, min_length=30, do_sample=False)[0]['summary_text'] for chunk in tqdm(chunks)]

# ✅ Run everything

full_text = extract_text_from_pdf(pdf_path)
chunks = chunk_text(full_text)
summaries = summarize_chunks(chunks)

# ✅ Final output
final_summary = "\n\n".join(summaries)
print("\n✅ SUMMARY DONE:\n")
print(final_summary)
```
